# Legends

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,x, label="x vs x") 
# mark a label to be able to connect this label 
# later to leged call

# lets plot again on the same axes
axes.plot(x,x**2,label= "x vs x^2")

# legend call
axes.legend(loc=0);

# loc - tell where to place the kegend
# 0 - "best location"
```

```python
fig = plt.figure(dpi =300)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x, label="x vs x") 

axes.plot(x,x**2,label= "x vs x^2")

# legend call
axes.legend(loc=(1.1,0.4));
# may call it ouside
# specify manually
# the same tuple as in axes
```

## color

```python
fig = plt.figure(dpi =300)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label="x",color="red");

ax.legend(loc=(1,0.5))
```

```python
# may pick HEX COLOR RGB code

fig = plt.figure(dpi =300)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label="x",color="#650bdb")
# obtained via google search
ax.plot(x,x+1,label="y",color="g")

ax.legend(loc=(1.1,0.5));
```

## line with style

```python
# choose the LINE WIDTH

fig = plt.figure(dpi =300)

ax = fig.add_axes([0,0,1,1])

# lw is a ratio of 1 = 100% of original size like above
# can be inserted as line width
ax.plot(x,x,label="x",color="#650bdb" ,  lw = 10)

# original 1 ration for comparison
ax.plot(x,x+1,label="y",color="g")



ax.legend(loc=(1.1,0.5));
```

## Line styles

```python
# possible line styles '--', '-', '-.', ':', "steps"



fig = plt.figure(dpi =300)

ax = fig.add_axes([0,0,1,1])


ax.plot(x,x-1,label="x1", ls="--",lw=2)
ax.plot(x,x,label="x2",color="#650bdb",lw = 2, linestyle = "-")
ax.plot(x,x+1,label="x3",color="b", ls = "-.", lw=2)
ax.plot(x,x+2,label="x4",color="g", linestyle = ":",lw=2)


ax.legend(loc=(1.1,0.5))
```

### Custom dash


```python
fig = plt.figure(dpi =300)

ax = fig.add_axes([0,0,1,1])


lines = ax.plot(x,x,label="x2",color="#650bdb",lw = 2, linestyle = "-")



ax.legend(loc=(1.1,0.5))
```

```python
fig = plt.figure(dpi =300)

ax = fig.add_axes([0,0,1,1])


lines = ax.plot(x,x,label="x2",color="#650bdb",lw = 2, linestyle = "-")
lines[0].set_dashes([5,2,10,2,30,2])
# 5-len of dashes
# 2-len of spaces
# 10-len of dash

# basically you define the length of a dash and a space after it

ax.legend(loc=(1.1,0.5))

# to grab line object

```

## Markers

```python
x
# is what we plot
# it is an array 
# contains points of plotting
```

```python
# if we want to plot all the point that occure in the graph

fig = plt.figure(dpi =300)
ax = fig.add_axes([0,0,1,1])

# markers have their own symbols
ax.plot(x,x,label="x2",color="red",lw = 2, marker ="o", ls='--', ms =10)
# may specify marker size
ax.plot(x,x+1,label="x2",color="#650bdb",lw = 2, marker ="X", ls='--', markersize=10)


ax.legend(loc=(1.1,0.5));
```

## EDIT market

```python
# able to play with markets

fig = plt.figure(dpi =300)
ax = fig.add_axes([0,0,1,1])

# markers properties
ax.plot(x,x,label="x1",color="red",lw = 2, marker ="s", ls='--', ms =10,
       markerfacecolor='red',markeredgewidth=4,markeredgecolor="orange")

ax.plot(x,x+1,label="original",color="green",lw = 2, marker ="s", ls='--', ms=10)
ax.legend(loc=(1.1,0.5));

```
