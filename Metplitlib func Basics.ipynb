```python
# since we work with pyplot portion fo the libriary

import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
# now we have a linear relationship

x
```

```python
y
```

```python
plt.plot(x,y)
plt.show() # to show the visualization
```

## Labeling Plots

```python
### TITLE AND AXIS

# all goes in the same cell

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("STRING TITLE"); # we only see render visualization with ;
# the same as plt.plot()
```

```python
### X and Y limits

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim(0,6)
plt.ylim(0,12)
plt.title("STRING TITLE");
```

## Save a Plot

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xlim(0,6)
plt.ylim(0,12)
plt.title("STRING TITLE")
plt.savefig("my_first_plot.png") # provide a filepath
```
