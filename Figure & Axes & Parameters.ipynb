```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# DATA
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

# 1 Create object

```python
fig = plt.figure()

# below you will find a report with some info on figure
# size and axes
# we cant see it now
```

# 2 Add axes

```python
axes = fig.add_axes([0,0,1,1])
```

# 3 Plot on set of axes

```python
axes.plot(x,y)
```

```python
### run all in one cell


fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b)
plt.show()
```

### example 1

```python
fig = plt.figure()

# LARGE axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

# small axes
axes2 = fig.add_axes([0.1,0.4,0.4,0.4])
axes2.plot(x,y)

plt.show()
```

## Naming Indevidual axis

```python
fig = plt.figure()

# LARGE axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

# labeling
axes1.set_xlim(0,8)
axes1.set_ylim(0,3500)
axes1.set_xlabel("X-AXES A")
axes1.set_ylabel("Y-AXES B (X**4)")
axes1.set_title("BIG GRAPH")



# SMALL axes
axes2 = fig.add_axes([0.3,0.4,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel("x-axes A")
axes2.set_ylabel("y-axes B")
axes2.set_title("small graph")



plt.show() #or use ; at the end to show the graph
```

# Parameters: DPI
dpi - dots per inch

```python
fig = plt.figure(dpi =300)
# makes the figure more visable
# dont make it too high to avoid ram overload

axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b);
```

# Parameters: Figure size

```python
fig = plt.figure(figsize=(2,1),dpi =300)
# figure size comes in inches (width-high)
# the figure would look big as soon as you open up
# the neccessary file


axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b);

fig.savefig("new_figure.png",bbox_inches="tight") # to reflect
# axes in the saved file
```
