```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a**4
```

```python
x = np.arange(0,10)
y = 2*x
```

# Single Demention

```python
fig,axes = plt.subplots()
# tuple unpacking and special formula to plon in the values

# the below cell will influence the above tuple
# and will plot x and y on axes in subplots
axes.plot(x,y);

# axes is a special plot object 
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)

# axis will become an array as soon as we add anothe column
# since it became an array we need
# to index the axes to adress and use plot function 

axes[0].plot(x,y)
axes[1].plot(a,b);
```

```python
axes
```

# Multiple Demention

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

# now axes will become multi-level array
# adress array[0] - rows with multiple axes 
# adress array[0][0] - axes within those rows

axes[0][0].plot(x,y)
axes[0][1].plot(a,b);

```

```python
axes
```

# Layout

```python
# here is one dementional array

fig, axes = plt.subplots(nrows=3,ncols=1)

axes[0].plot(x,y)

# to make sure no axes and tikers are overlapping
# run at the end of the script
plt.tight_layout();
```

## Manually

```python
fig, axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(x,y)
axes[1][1].plot(x,y)

# we are able to manually adjust the preview
fig.subplots_adjust(wspace=0.3, hspace=0.3)

# both wspace,hspace are expressced as fractions from width and hight
# 30% of the average w and h 
```

## Parameters SubPlots

```python
fig, axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][0].set_xlabel("X [0,0]")

axes[0][1].plot(x,y)
axes[0][1].set_ylabel("Y [0,1]")

axes[1][0].plot(x,y)
axes[1][0].set_title ("title [1,0]")
axes[1][0].set_ylim(0,10)

axes[1][1].plot(x,y)
axes[1][1].set_xlim(0,4)

# we are able to manually adjust the preview
plt.tight_layout()
```

## TITLE THE ENTIRE PLOT

```python
fig, axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][0].set_xlabel("X [0,0]")

axes[0][1].plot(x,y)
axes[0][1].set_ylabel("Y [0,1]")

axes[1][0].plot(x,y)
axes[1][0].set_title ("title [1,0]")
axes[1][0].set_ylim(0,10)

axes[1][1].plot(x,y)
axes[1][1].set_xlim(0,4)

fig.suptitle("FIGURE LEVEL TITLE", fontsize = 30) # caption level


# we are able to manually adjust the preview
plt.tight_layout()
```
